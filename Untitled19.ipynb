```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from wordcloud import WordCloud
from google.colab import files
```

```python
# Step 2: Upload CSV file
uploaded = files.upload()
```

```python
# Step 3: Load dataset
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename, encoding='latin-1')
print(" Dataset Loaded Successfully!\n")
print(" Preview:")
display(df.head())
```

```python
# Step 4: Basic info
print("\nDataset Info:")
df.info()
```

```python
# Step 5: Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 6: Show unique sentiment labels
print("\nSentiment Classes:")
print(df['sentiment'].value_counts())
```

```python
# Step 7: Visualize class distribution
plt.figure(figsize=(5,4))
sns.countplot(x='sentiment', data=df, palette='coolwarm')
plt.title("Sentiment Distribution")
plt.show()
```

```python
# Step 8: Clean text data
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def clean_text(text):
  text = str(text).lower()
  text = re.sub(r'http\S+', '', text)
  text = re.sub(r' [^a-z\s]', '', text)
  # remove links
  # remove punctuation/numbers
  text = " ".join([word for word in text.split() if word not in stop_words])
  return text
df['clean_text'] = df['text'].apply(clean_text)
print("\n Cleaned Text Sample:")
display (df[['text', 'clean_text']].head())
```

```python
# Step 9: WordCloud for visualization
text_all = " ".join(df['clean_text'])
plt.figure(figsize=(10,6))
wordcloud = WordCloud (width=800, height=400, background_color='white').generate (text_all)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Common Words in Dataset")
plt.show()
```

```python
# Step 10: Split data
df_cleaned = df.dropna(subset=['sentiment'])
x = df_cleaned['clean_text']
y = df_cleaned['sentiment']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 11: Convert text to numerical features
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 12: Train Logistic Regression model
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

#Step 13: Evaluate
y_pred = model.predict(X_test_vec)
```

```python
# Step 13: Evaluate
y_pred = model.predict(X_test_vec)
print("\n Model Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
# Step 14: Confusion Matrix
plt.figure(figsize=(5,4))
```
